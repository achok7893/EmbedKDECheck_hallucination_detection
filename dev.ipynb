```python
import json
import copy
# Open the file and read line by line
file_path = r"./data/a_raw/qags/mturk_xsum.jsonl"
data = []

with open(file_path, "r", encoding="utf-8") as file:
    for line in file:
        data.append(json.loads(line.strip()))  # Parse each line as a JSON object

i = 0
articles = []
responses = []
l_nb_yes = []
l_nb_no = []

for i in range(len(data)):
       
    article = data[i]["article"]
    sentence = data[i]["summary_sentences"][0]["sentence"]
    nb_yes = 0
    nb_no = 0
    for iter_ in data[i]["summary_sentences"][0]["responses"]:
        if iter_['response'] == "yes":
            nb_yes += 1
        elif iter_['response'] == "no":
            nb_no+=1
    
    articles.append(article)
    responses.append(sentence)
    l_nb_yes.append(nb_yes)
    l_nb_no.append(nb_no)
```

```python
hall_scores = []
l_results = []
for article, response in zip(articles, responses):
    res = embed2kde.get_scores_from_input_output_texts(article, response, bert_model)
    hall_scores.append(res["max_hallucination_score"])
    l_results.append(copy.deepcopy(res))
```

```python
import numpy as np
from scipy.stats import spearmanr


# Compute Spearman's rank correlation
corr, p_value = spearmanr(hall_scores, l_nb_no)

print(f"Spearman Correlation: {corr:.4f}")
print(f"P-value: {p_value:.4f}")
```
